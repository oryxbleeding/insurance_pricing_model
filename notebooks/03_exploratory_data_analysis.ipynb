```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('../data/processed/insurance_pricing_data_processed.csv')
numeric_vars = ['Age', 'Income', 'Income_per_Capita', 'Vehicle_Age', 'Mileage', 'Claims', 'Current_Premium', 'Regional_Claim_Rate', 'Economic_Index']

for var in numeric_vars:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[var], kde=True)
    plt.title(f'Verteilung von {var}')
    plt.show()
```

```python
# Scatterplot von Income vs. Tariff
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Income'], y=data['Tariff'])
plt.title('Income vs. Tariff')
plt.show()
```

```python
# Tariff nach Vehicle_Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Vehicle_Type_Sports Car', y='Tariff', data=data)
plt.title('Tariff nach Fahrzeugtyp (Sports Car)')
plt.show()
```

```python
# Korrelationsmatrix berechnen
corr_matrix = data.corr()

# Heatmap plotten
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Korrelationsmatrix der Features')
plt.show()
```
